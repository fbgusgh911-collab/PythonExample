```python
# 8. 특정 위치에 삽입 (insert) [cite: 397]
thislist = ["apple", "banana", "cherry"]
thislist.insert(1, "orange")
print(thislist)
```

```python
# 10. 마지막 항목 제거 (pop) [cite: 404]
thislist = ["apple", "banana", "cherry"]
thislist.pop()
print(thislist)
```

```python
# 11. 리스트 반복문 (for loop) [cite: 406]
thislist = ["apple", "banana", "cherry"]
for x in thislist:
    print(x)
```

```python
# 12. 리스트 컴프리헨션 [cite: 415]
fruits = ["apple", "banana", "cherry", "kiwi", "mango"]
newlist = [x for x in fruits if "a" in x]
print(newlist)
```
