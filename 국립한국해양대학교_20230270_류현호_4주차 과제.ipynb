```python
# [Section 2] Built-in Modules (datetime, math, json) [cite: 43, 52, 57]

import datetime
import math
import json
```

```python
# [Section 3] Regular Expressions (RegEx) [cite: 380, 384]

import re

txt = "The rain in Spain"
```

```python
# findall: 모든 일치 항목 찾기 [cite: 481, 482]
matches = re.findall("ai", txt)
print(f"Findall 'ai': {matches}") # ['ai', 'ai'] [cite: 486]
```

```python
# search: 첫 번째 일치 항목 찾기 [cite: 504, 505]
x = re.search("\s", txt) # 첫 공백 위치 찾기 [cite: 509]
print(f"First space position: {x.start()}") # [cite: 510]
```

```python
# split & sub: 분할 및 교체
split_txt = re.split("\s", txt) # 공백 기준 분할
sub_txt = re.sub("\s", "_", txt) # 공백을 언더바로 교체
print(f"문자열 분할: {split_txt}")
print(f"문자열 교체: {sub_txt}")
print("")
```

```python
# F-String을 이용한 정밀한 포맷팅
price = 59
tax = 0.25
total = price + (price * tax)
print(f"최종 가격(소수점 2자리): ${total:.2f}")
```
