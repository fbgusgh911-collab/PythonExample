```python
# 1. 데이터 생성 (1부터 100까지의 숫자가 들어있는 리스트)
data_list = list(range(1, 101))
total_count = len(data_list)
```

```python
# 2. 실행 목표 설정 (전체의 33%)
target_percentage = 0.33
limit = int(total_count * target_percentage)

print(f"전체 데이터 개수: {total_count}개")
print(f"실행 목표(33%): {limit}개\n")
```

```python
# 3. 반복문과 조건문을 이용한 실행
print("--- 실행 결과 ---")
executed_count = 0

for data in data_list:
    # 현재 데이터가 33번째 이내인 경우에만 실행
    if data <= limit:
        # 강의 자료의 예시(a=33, b=200)를 응용하여 출력 로직 구성
        a = data
        b = 100 # 기준값
        
        if b > a:
            print(f"[{data}번째 실행] {a}은(는) {b}보다 작으므로 조건 충족 (33% 범위 내)")
        
        executed_count += 1
    else:
        # 33%를 초과하면 반복 종료
        break

print("\n--- 최종 보고 ---")
print(f"총 {executed_count}개의 데이터가 성공적으로 처리되었습니다.")
```
