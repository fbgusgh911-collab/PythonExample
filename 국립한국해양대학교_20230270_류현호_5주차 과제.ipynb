```python
# [Python08] 클래스 개념 활용
class DataProcessor:
    def __init__(self, name, total_count):
        self.name = name
        self.total_count = total_count
        self.data = list(range(1, total_count + 1))
        self.limit = int(total_count * 0.33)  # 33% 계산

    def process_33_percent(self):
        """상위 33% 데이터를 리스트로 반환하는 메서드"""
        print(f"[{self.name}] 데이터 처리를 시작합니다. (목표: {self.limit}개)")
        return self.data[:self.limit]
```

```python
class Person:
  def __init__(self, name):
    self.name = name
  def printname(self):
    print(self.name)
p1 = Person("Tobias")
p2 = Person("Linus")
p1.printname()
p2.printname()
```

```python
class Calculator:
  def add(self, a, b):
    return a + b
  def multiply(self, a, b):
    return a * b
calc = Calculator()
print(calc.add(5, 3))
print(calc.multiply(4, 7))
```
